```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select

# initialising our driver and getting the url to the webpage

driver=webdriver.Chrome()
driver.get('https://dharani.telangana.gov.in/knowLandStatus')
```

```python
# ensuring the search by survey number button is on

x=driver.find_element(By.ID,'survyno')
if x.is_selected():
    pass
else:
    x.click()
```

```python
# choose the district value
district=Select(driver.find_element(By.ID,'districtID'))
district.select_by_value("13")
```

```python
# choose the mandal value
mandal=Select(driver.find_element(By.ID,'mandalID'))
mandal.select_by_value("261")
```

```python
# chose the village value
village=Select(driver.find_element(By.ID,'villageId'))
village.select_by_value("1301023")
```

```python
# get the survey number list
Survey_no_list=Select(driver.find_element(By.ID,'surveyIdselect'))
```

```python
# storing all the elements as a list
a=[]
for option in Survey_no_list.options:

    a.append(option.text)
    
print(a)
```

```python
# function to get the survey list for each combination of district,mandal,village

def get_survey_list(district_value,mandal_value,village_value): 

    district=Select(driver.find_element(By.ID,'districtID'))
    district.select_by_value(district_value)

    mandal=Select(driver.find_element(By.ID,'mandalID'))
    mandal.select_by_value(mandal_value)

    village=Select(driver.find_element(By.ID,'villageId'))
    village.select_by_value(village_value)

    Survey_no_list=Select(driver.find_element(By.ID,'surveyIdselect'))

    element_survey_list=[]

    for option in Survey_no_list.options:

        element_survey_list.append(option.text)

    return element_survey_list

survey_list = get_survey_list("13","261","1301023")

print(survey_list)

```
